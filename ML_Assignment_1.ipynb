<b>1. What does one mean by the term &quot;machine learning&quot;?

Machine learning is a process by which machines can solve problems without being explicitly programmed. Suppose there is a task T and a machine is trained which gives it experience E and we measure the performance of the machine on T using a metric M. If the metric M improves on task T with experience E. Then we say machine is learning to solve a problem

<b>18.What precisely is the train-dev kit, when will you need it, how do you put it to use?

Train-dev set is a part of training data kept aside to decide whether a model's poor performance on validation data is due to overfitting or data mismatch. 
- If a model performs well on training data and train-dev set but poorly on validation set. It means there is data mismatch
- If a model performs well on training data and poorly train-dev set and validation set. It means there is overfitting

<b>19.What could go wrong if you use the test set to tune hyperparameters?
